```python
#lesson 3
# Даны значения зарплат из выборки выпускников: 
# 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150.
# Посчитать (желательно без использования статистических методов наподобие std, var, mean) 
# среднее арифметическое, 
# среднее квадратичное отклонение, 
# смещенную и несмещенную оценки дисперсий для данной выборки.
import numpy as np
import math

from math import factorial
import math
def combinations(n,k):
    return int(factorial(n)/(factorial(k)*factorial(n-k)))

a = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150], int)
print(a)
print(f"sum(a) = {sum(a)} " )
print(f" len(a) = {len(a)} " )
vaverage = sum(a)/len(a)
print(f" среднее арифметическое average(a) = {vaverage} " )

vstddevi = 0
for i in a:
    vstddevi = (i - vaverage)**2+vstddevi
    
vstddeviE = math.sqrt(vstddevi/(len(a)-1))
# print(i)
print(f" среднее квадратичное отклонение  standart deviation a {vstddeviE}")
```

```python
# смещенную и несмещенную оценки дисперсий для данной выборки.
shiftDisp = vstddevi/len(a)
print(shiftDisp)
```
